## 2. Detecci√≥n de Sentimiento y Clasificaci√≥n Zero-Shot:

```python
# Importamos pipeline de la librer√≠a Transformers
from transformers import pipeline
```

2.1 ü§ó Utilizamos Transformers de Hugging Face

```python
# Cargamos el pipeline de an√°lisis de sentimientos con el modelo BETO, que es un modelo BERT entrenado con un gran corpus de espa√±ol
clasificador = pipeline("sentiment-analysis", model="finiteautomata/beto-sentiment-analysis")
```

```python
# Funci√≥n para procesar los tickets y analizar sentimiendo usando Transformers
def analizar_sentimiento_transformers(texto):
  if not clasificador:
    return []

  sentimientos = clasificador(texto)

  try:
    # Formatear resultados
    resultados = []
    for sent in sentimientos:
      resultados.append({
            'etiqueta': sent['label'],
            'confianza': round(sent['score'], 2),
            'ticket': texto
                    })
    return resultados
  except Exception as e:
    return f"‚ùå Error: {e}"

if clasificador:
  # Aplicamos la funci√≥n
  print(" üîé Analizando Sentimiento de los Tickets...")
  print(' ')
  sentimientos_list = []
  for ticket in df['tickets'].tolist():
    sentimientos_list.extend(analizar_sentimiento_transformers(ticket))

  for sent in sentimientos_list:
    print(f"  ‚Ä¢ Ticket:{sent['ticket']} ‚Üí {sent['etiqueta']} (confianza: {sent['confianza']})")
```

```python
# Cargamos el pipeline de clasificaci√≥n zero-shot con el modelo 'bert-base-spanish-wwm-cased-xnli'
zero_shot_clasificacion = pipeline("zero-shot-classification", model="Recognai/bert-base-spanish-wwm-cased-xnli")
```

```python
# Funci√≥n para categorizar los tickets usando Transformers
def analizar_categorias_transformers(texto, etiquetas_candidatas):
  if not zero_shot_clasificacion:
    return []

  categorias = zero_shot_clasificacion(texto, etiquetas_candidatas)
  try:
    # Formatear resultados
    resultados = []
    resultados.append({
          'categorias': categorias['labels'],
          'confianza': categorias['scores'],
          'ticket': texto
                  })
    return resultados

  except Exception as e:
    return f"‚ùå Error: {e}"

if zero_shot_clasificacion:
  print(" üîé Analizando categor√≠as de los Tickets...")
  print(' ')
  categorias_list = []
  etiquetas_candidatas = df['categorias'].unique().tolist()
  for ticket in df['tickets'].tolist():
    categorias_list.extend(analizar_categorias_transformers(ticket, etiquetas_candidatas))

  for cat in categorias_list:
    print(f"  ‚Ä¢ Ticket:{cat['ticket'][:50]} ‚Üí {cat['categorias'][0]} (confianza: {cat['confianza'][0]:.2f})")
```

*   An√°lisis de Sentimientos: Para dos casos, las clasificaciones difieren de la que realizamos manualmente. Esto puede notarse en los primeros tickets que se√±alamos como **NEGATIVOS** y el modelo los calific√≥ como **NEUTROS** con un puntaje mayor al 0.5.


*  Clasificaci√≥n Zero-Shot: Observamos que los primeros tres casos coinciden con lo calificado manualmente. Al cuarto ticket, el modelo lo categoriza como "elogio", cuando lo hemos etiquetado como "consulta de stock". Para el quinto y sexto ticket las categor√≠as vuelven a coincidir. En el s√©ptimo encontramos otra diferencia, ya que lo califica como "elogio" y fue cargado como "consulta de producto". Los siguientes dos tickets coinciden con las etiquetas manuales pero el d√©cimo se distingue de nuevo del modelo, siendo para √©ste un caso de "consulta de producto", pero identificado por nosotros como "producto defectuoso". Los siguientes tres tickets tambi√©n se ajustan a nuestra categorizaci√≥n, ya que los detecta como "elogio". Pero para los √∫ltimos dos la clasificaci√≥n es la inversa a la realizada manual. El ticket 14 lo se√±ala como "consulta de producto" cuando lo cargamos como "elogio" y el ticket 15, lo opuesto. Es detectado por el modelo como "elogio", cuando fue cargado en nuestro dataset como "consulta de producto".

## 3. Extracci√≥n de Informaci√≥n Clave (NER/QA):

3.1 ü§ó Reconocimiento de entidades nombradas (NER) Y QA con Hugging Face

```python
# Cargamos el pipeline y el modelo de reconocimiento de entidades nombradas (NER) en espa√±ol
ner_nlp = pipeline("ner", model="mrm8488/bert-spanish-cased-finetuned-ner",  aggregation_strategy="simple")
```

```python
# Funci√≥n para procesar los tickets y extraer entidades usando Transformers
def analizar_entidades_transformers(texto):
  if not ner_nlp:
    return []
  entidades = ner_nlp(texto)

  try:
    # Formatear resultados
    resultados = []
    for ent in entidades:
      resultados.append({
            'texto': ent['word'],
            'etiqueta': ent['entity_group'],
            'confianza': round(ent['score'], 2),
            'posicion': (ent['start'], ent['end'])
        })
    return resultados
  except Exception as e:
    return f"‚ùå Error: {e}"

if ner_nlp:
  # Aplicamos la funci√≥n
  print(" üîé Analizando tickets...")
  print("\nüìä Entidades encontradas:")
  entidades_encontradas_list = []
  for ticket in df['tickets'].tolist():
    entidades_encontradas_list.extend(analizar_entidades_transformers(ticket))
  for ent in entidades_encontradas_list:
    print(f"  ‚Ä¢ {ent['texto']} ‚Üí {ent['etiqueta']} (confianza: {ent['confianza']})")
```

```python
# Cargamos pipeline y modelo de Question Answering de respuestas a preguntas en espa√±ol
qa_nlp = pipeline("question-answering", model="PlanTL-GOB-ES/roberta-large-bne-sqac")
```

```python
# Funci√≥n para para obtener una pregunta autom√°tica por ticket con Transformers
def obtener_respuestas_qa(df, qa_nlp, pregunta="¬øCu√°l es el problema del cliente?"):
    """
    Procesa los tickets del DataFrame y responde a la pregunta especificada usando el modelo QA.

    Par√°metros:
    - df: DataFrame que contiene una columna 'tickets'
    - modelo_qa: pipeline QA de Hugging Face
    - pregunta: string con la pregunta a responder sobre cada ticket

    Retorna:
    - DataFrame con columnas 'ticket', 'answer' y 'score'
    """
    resultados = []
    for ticket in df['tickets'].tolist():
        try:
            salida = qa_nlp(question=pregunta, context=ticket)
            resultados.append({
                'ticket': ticket[:50] + '...',
                'answer': salida['answer'],
                'score': round(salida['score'], 3)
            })
        except Exception as e:
            resultados.append({
                'ticket': ticket[:50] + '...',
                'answer': 'Error al procesar',
                'score': None
            })
            print(f"Error con ticket: {ticket[:30]}... -> {e}")

    return pd.DataFrame(resultados)

respuestas_df = obtener_respuestas_qa(df, qa_nlp)
display(respuestas_df[['ticket', 'answer', 'score']])
```

3.2 ‚ú® Reconocimiento de entidades nombradas (NER) y QA con Gemini API

```python
# Funci√≥n para reconocimiento de entidades usando Gemini
def extraer_entidades_ner_gemini(df, cliente, modelo):
    """
    Extrae entidades nombradas del texto de los tickets.

    Par√°metros:
    - df: DataFrame con una columna 'tickets'
    - cliente: cliente Gemini autenticado
    - modelo: nombre del modelo a usar (ej. 'gemini-2.0-flash')

    Retorna:
    - String con las entidades extra√≠das y clasificadas por tipo
    """
    if not cliente:
        return "‚ùå Cliente Gemini no disponible."

    lista_tickets = df['tickets'].dropna().tolist()

    prompt = f"""
Extra√© todas las entidades nombradas del siguiente texto en espa√±ol argentino y clasific√°las:

CATEGOR√çAS:
- PERSONA: Nombres de personas
- LUGAR: Ciudades, pa√≠ses, barrios, direcciones, lugares espec√≠ficos
- ORGANIZACI√ìN: Empresas, universidades, instituciones
- MISCEL√ÅNEO: Otros nombres propios (productos, eventos, marcas)

FORMATO DE RESPUESTA:
[ENTIDAD] ‚Üí [CATEGOR√çA] ‚Üí [BREVE EXPLICACI√ìN]

TEXTO A ANALIZAR:
{lista_tickets}
"""

    try:
        respuesta = cliente.models.generate_content(
            model=modelo,
            contents=[prompt]
        )
        return respuesta.text
    except Exception as e:
        return f"‚ö†Ô∏è Error al generar contenido con Gemini: {e}"
```

```python
# Funci√≥n para realizar pregunta autom√°tica con Gemini
def obtener_respuesta_gemini(texto):
  if not cliente:
    return " ‚ùå Cliente Gemini no disponible"
  tickets = df['tickets'].tolist()
  for ticket in df['tickets'].tolist():
    prompt_qa = "¬øCu√°l es el problema principal que se extrae de cada ticket"
    contexto = tickets

    prompt_qa = f"""Respond√© la siguiente pregunta basada en el contenido de los tickets:
               Pregunta: {prompt_qa}
                Texto: {tickets}

                  """

  try:
      respuesta_qa = cliente.models.generate_content(
      model=modelo,
      contents=[prompt_qa]
      )
      return respuesta_qa.text


  except Exception as e:
    return f"‚ùå Error: {e}"


if cliente:
    print("üîé Analizando con Gemini...")
    resultado_gemini = obtener_respuesta_gemini(df['tickets'].tolist())
    print("\nAn√°lisis de Gemini:")
    print(resultado_gemini)
else:
    print("‚è≠Ô∏è  Saltando an√°lisis con Gemini (API Key no disponible)")
```

## 4. Generaci√≥n de Respuesta Personalizada

4.1 ü§ó Respuesta personalizada con Hugging Face

```python
# Cargamos pipeline y modelo para generar texto
generador = pipeline("text-generation", model="datificate/gpt2-small-spanish")
```

```python
def generar_respuestas_hf(df, generador):
    """
    Genera respuestas autom√°ticas personalizadas para cada ticket del DataFrame
    seg√∫n su etiqueta de sentimiento ('negativo', 'neutro', 'positivo') utilizando Hugging Face.

    Par√°metros:
    - df: DataFrame con columna 'etiquetas'
    - generador: pipeline de Hugging Face para generaci√≥n de texto

    Retorna:
    - DataFrame con una columna 'respuesta' correspondiente a cada etiqueta
    """
    respuestas = []

    for etiqueta in df['etiquetas'].tolist():
        if etiqueta == 'negativo':
            prompt = ("Escribe una respuesta de no m√°s de 4 l√≠neas para atender la solicitud o queja del cliente, "
                      "con tono agradable. La misma debe comenzar con un pedido de disculpas e indicar que soporte se "
                      "comunicar√° a la brevedad si el problema es por env√≠os o desde el √°rea de devoluciones si el problema es por cambio de producto.")
        elif etiqueta == 'neutro':
            prompt = ("Escribe una respuesta de no m√°s de 4 l√≠neas para atender la solicitud del cliente. "
                      "Se debe comunicar, con tono agradable, que atenci√≥n al cliente se comunicar√° a la brevedad para resolver su consulta.")
        else:
            prompt = ("Escribe una respuesta para agradecer el comentario y la confianza, "
                      "mencionando que su apoyo es la motivaci√≥n para seguir creciendo. "
                      "La misma no debe extenderse m√°s de 4 l√≠neas.")

        try:
            output = generador(prompt, max_new_tokens=100, truncation=True)
            respuesta = output[0]['generated_text'].strip()
        except Exception as e:
            respuesta = f"‚ùå Error: {e}"

        respuestas.append(respuesta)

    return pd.DataFrame({'respuesta': respuestas})
```

```python
respuestas_cliente = generar_respuestas_hf(df, generador)
print(respuestas_cliente.head())
```

4.2 ‚ú® Respuesta Personalizada con Gemini

> Gemini API tiene un l√≠mite de n√∫mero de pedidos de `generate_content` que se p√∫eden realizar por minuto en la versi√≥n gratuita de ` gemini-2.0-flash model`. Para evitar que arroje un error, cargaremos la librer√≠a time que nos permite pausar la ejecuci√≥n.





```python
import time
import pandas as pd

def generar_respuestas_gemini(df, cliente, modelo, pausa=10):
    """
    Genera respuestas personalizadas para cada ticket usando Gemini API.

    Par√°metros:
    -----------
    df : pd.DataFrame
        DataFrame que debe contener una columna 'etiquetas' con valores 'negativo', 'neutro' o 'positivo'.

    cliente : object
        Cliente Gemini ya autenticado, por ejemplo: cliente = genai.GenerativeModel(...)

    modelo : str, opcional
        Nombre del modelo a usar (por defecto 'gemini-1.5-flash-latest').

    pausa : int, opcional
        Segundos de espera entre cada solicitud para evitar rate limit (por defecto 10).

    Retorna:
    --------
    pd.DataFrame
        Un DataFrame con una columna de respuestas generadas.
    """
    respuestas = []

    for etiqueta in df['etiquetas'].tolist():
        etiqueta = str(etiqueta).lower()

        if etiqueta == 'negativo':
            respuesta_inicial = "Estimado/a cliente, lamentamos los inconvenientes ocasionados con su pedido. "
            prompt = f"""Redact√° una respuesta del servicio de atenci√≥n al cliente que comience as√≠:
{respuesta_inicial}
El mensaje debe ser en tono agradable y debe indicar que si es por problemas de env√≠os el √°rea de soporte le solicitar√° a la brevedad n√∫mero
de orden para su seguimiento y resoluci√≥n o si quiere realizar un cambio, el √°rea de devoluciones se comunicar√°.
No debe extenderse m√°s de 4 l√≠neas.
"""
        elif etiqueta == 'neutro':
            respuesta_inicial = "Estimado/a cliente, gracias por escribirnos."
            prompt = f"""Redact√° una respuesta del servicio de atenci√≥n al cliente que comience as√≠:
{respuesta_inicial}
El mensaje debe ser en tono agradable y debe indicar que a la brevedad se comunicar√° el equipo de soporte
para responder su consulta. No debe extenderse m√°s de 3 l√≠neas.
"""
        else:
            respuesta_inicial = "Estimado/a cliente, agradecemos tu comentario. Tu apoyo nos motiva."
            prompt = f"""Redact√° una respuesta del servicio de atenci√≥n al cliente que comience as√≠:
{respuesta_inicial}
Es un mensaje de agradecimiento al cliente por su comentario positivo.
No debe extenderse m√°s de 3 l√≠neas.
"""

        try:
            respuesta = cliente.models.generate_content(
                model=modelo,
                contents=[prompt]
            )
            respuestas.append(respuesta.text.strip())
        except Exception as e:
            respuestas.append(f"[ERROR]: {str(e)}")

        time.sleep(pausa)

    return pd.DataFrame({'respuesta': respuestas})
```

```python
respuestas_g = generar_respuestas_gemini(df, cliente, modelo)
respuestas_cliente_gemini= pd.DataFrame(respuestas_g)
respuestas_cliente_gemini
```

```python
# Importamos Gradio para crear una interfaz
import gradio as gr
```

```python
# Definimos las funciones para analizar el ingreso de un Ticket de Soporte a la interfaz Gradio
def analisis_sentimiento_gemini(texto):
   if not cliente:
    return " ‚ö† Cliente Gemini no disponible"

   try:
    pregunta = f"""Clasific√° el siguiente corpus de tickets como positivo, negativo o neutral:
    Texto: {texto}
    """

    respuesta_sentimiento = cliente.models.generate_content( model=modelo,
                                                            contents=[pregunta])
    return respuesta_sentimiento.text

   except Exception as e:
    return f"‚ùå Error: {e}"
```

```python
def analisis_categoria_gemini(texto):
  if not cliente:
    return " ‚ö† Cliente Gemini no disponible"

  pregunta_zs = f"""Clasific√° el corpus de tickets en una de estas categor√≠as: {df['categorias'].unique().tolist()}.

                    Texto: {texto}

                    """
  try:
      respuesta_zs = cliente.models.generate_content(
                     model=modelo,
                     contents=[pregunta_zs]
                      )
      return respuesta_zs.text


  except Exception as e:
    return f"‚ùå Error: {e}"
```

```python
def analisis_entidades_gemini(texto):
  if not cliente:
    return " ‚ö† Cliente Gemini no disponible"
  prompt_ner = f"""Extra√© todas las entidades nombradas del siguiente texto (personas, organizaciones, lugares, objetos) y clasific√°las:
                Texto: {texto}
                """
  try:
    respuesta_ner = cliente.models.generate_content( model=modelo,
                                                    contents=[prompt_ner]
                                                     )

    return respuesta_ner.text


  except Exception as e:
    return f"‚ùå Error: {e}"
```

```python
def respuesta_clave_gemini(texto):
  if not cliente:
    return " ‚ö† Cliente Gemini no disponible"
  prompt_qa = "¬øCu√°l es el problema principal que se extrae de cada ticket"
  contexto = texto

  prompt_qa = f"""Respond√© la siguiente pregunta basada en el contenido de los tickets:
  Pregunta: {prompt_qa}
  Texto: {texto}

  """
  try:
    respuesta_qa = cliente.models.generate_content(
      model=modelo,
      contents=[prompt_qa]
  )
    return respuesta_qa.text


  except Exception as e:
    return f"‚ùå Error: {e}"
```

```python
def respuesta_generada_gemini(texto):
  if not cliente:
    return " ‚ö† Cliente Gemini no disponible"
  etiqueta = analisis_sentimiento_gemini(texto)
  if etiqueta == 'negativo':
    respuesta_inicial = "Estimado/a cliente, lamentamos los inconvenientes ocasionados con su pedido. "
    prompt = f"""Redact√° una respuesta del servicio de atenci√≥n al cliente que comience as√≠:{respuesta_inicial}
    El mensaje debe ser en tono agradable y debe indicar que si es por problemas de env√≠os el √°rea de soporte le solicitar√° a la brevedad n√∫mero
    de orden para su seguimiento y resoluci√≥n o si quiere realizar un cambio, el √°rea de devoluciones se comunicar√°.
    No debe extenderse m√°s de 4 l√≠neas.
    """
  elif etiqueta == 'neutro':
    respuesta_inicial = "Estimado/a cliente, gracias por escribirnos."
    prompt = f"""Redact√° una respuesta del servicio de atenci√≥n al cliente que comience as√≠:
    {respuesta_inicial}
    El mensaje debe ser en tono agradable y debe indicar que a la brevedad se comunicar√° el equipo de soporte
para responder su consulta. No debe extenderse m√°s de 3 l√≠neas.
"""
  else:
            respuesta_inicial = "Estimado/a cliente, agradecemos tu comentario. Tu apoyo nos motiva."
            prompt = f"""Redact√° una respuesta del servicio de atenci√≥n al cliente que comience as√≠:
{respuesta_inicial}
Es un mensaje de agradecimiento al cliente por su comentario positivo.
No debe extenderse m√°s de 3 l√≠neas.
"""

  try:
    respuesta = cliente.models.generate_content(
                model=modelo,
                contents=[prompt]
            )
    return respuesta.text.strip()
  except Exception as e:
    (f"[ERROR]: {str(e)}")

  time.sleep(20) # Pausa de 20 segundos
```

```python
"""" Esta funci√≥n permite ingresar un nuevo ticket de soporte y devuelve:
       ‚Ä¢ Sentimiento predicho del ticket
       ‚Ä¢ Categor√≠a (Zero-Shot)
       ‚Ä¢ Entidad o Respuesta Clave
       ‚Ä¢ Respuesta del modelo de lenguaje  """

def interfaz_ticket_soporte(texto):
  if not texto:
    return " ‚ö†Ô∏è No se ingres√≥ ning√∫n ticket. Por favor ingresa uno: "

  if not cliente:
      return " ‚ùå Cliente Gemini no disponible. Por favor, aseg√∫rate de haber configurado tu API Key."

  sentimiento = analisis_sentimiento_gemini(texto)
  categoria = analisis_categoria_gemini(texto)
  entidades = analisis_entidades_gemini(texto)
  respuesta_clave= respuesta_clave_gemini(texto)
  respuesta_personalizada = respuesta_generada_gemini(texto)

  # Pasamos los resultados en un solo bloque
  resultados_str = f"An√°lisis del Ticket:\n\n" \
                   f"Sentimiento: {sentimiento}\n" \
                   f"Categor√≠a: {categoria}\n" \
                   f"Respuesta Clave: {respuesta_clave}\n" \
                   f"Entidades: {entidades}\n" \
                   f"Respuesta Personalizada: {respuesta_personalizada}"
  return resultados_str




```
